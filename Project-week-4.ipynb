# ROI calculator project
<p>make a return on investment calculator the class will be ROIcalculator with methods to define Monthly income, monthly expenses, what the total cash flow would be monthly so monthly inc - monthly exp.<p>
<p>Monthly income is rental/laundry/storage/misc income added up together<p>
<p>Monthly expenses include utility expenses like water, gas, electric, sewage, garbage, also HOA, Lawn/snow, vacancy, repair, CapEx, property management and mortgage.<p>
<p>To get annual income you will obtain the difference of month inc - month exp, then multiply by 12 for 12 months in a year.<p>
<p>Finally to get cash on cash return we define the down payment, + closing costs,+ rehab budget,+ misc costs to receive the TOTAL investment.<p>
    
<p>now we divide  our Annual Cash Flow with  total investment to get a decimal, multiply by 100 to get the percentage and now we have our ROI percent<p>

```python
class ROIcalc:
    
    def __init__(self, monthly_inc = 0, monthly_exp = 0, monthly_flow = 0, annual_flow = 0, investment = 0,
                cash_roi = 0):
        self.monthly_inc = monthly_inc
        self.monthly_exp = monthly_exp
        self.monthly_flow = monthly_flow
        self.annual_flow = annual_flow
        self.investment = investment
        self.cash_roi = cash_roi
    
    def monthly_income(self):
        '''Ask user for their monthly income and add it up together'''
        
        rental = int(input('What is your monthly rental income?'))
        laundry = int(input('What is your monthly laundry income?'))
        storage = int(input('What is your monthly storage income?'))
        misc = int(input('What is your monthly miscellanous income?'))
        
        self.monthly_inc = sum([rental, laundry, storage, misc])
        
        
        
    def monthly_expense(self):
        '''Ask user for monthly expenses'''
        
        tax = int(input('How much do you pay in taxes? '))
        insurance = int(input('How much is your insurance? '))
        util = input('Do you have utility expenses? "yes" or "no" ').lower()
        
        while util not in ['yes','no']:
            util = input('Please input "yes" or "no"').lower()
        if util == 'yes':
            elec = int(input('What is your monthly cost for electricity? '))
            water = int(input('What is your monthly cost for water? '))
            sewer = int(input('What is your monthly cost for sewer? '))
            garb = int(input('What is your monthly cost for garbage? '))
            gas = int(input('What is your monthly cost for gas? '))
        elif util == 'no':
            elec = 0 
            water = 0
            sewer = 0
            garb = 0
            gas = 0
        
        
        hoa = int(input('Home owners association fees? '))
        lawn = int(input('Monthly expense for lawn? '))
        snow = int(input('Monthly expense for snow? '))
        vacancy = int(input('Monthly expense for vacancy? '))
        repairs =  int(input('Monthly expense for repairs? '))
        capital_exp = int(input('Monthly expense for capital expenditures? '))
        property_manage = int(input('Monthly expense for property management? '))
        mortgage = int(input('Monthly expense for mortgage? '))
        
        self.monthly_exp = sum([tax, insurance, elec, water, sewer, garb, gas, hoa, lawn, snow, vacancy, 
                                repairs,capital_exp, property_manage, mortgage])
    
    def annual_cf(self):
        '''subtract monthly income and monthly exp to get cash flow, multiply cash flow by 12 to get
        annual cash flow'''
        
        self.monthly_flow = self.monthly_inc - self.monthly_exp
        self.annual_flow = self.monthly_flow * 12
        print(f'Your monthly cash flow is {self.monthly_flow} and your annual cash flow is {self.annual_flow}')
        
    
    
    def total_investment(self):
        '''getting total investment'''
        down_payment = int(input('How much was your down payment? '))
        closing_costs = int(input('How much was your closing costs? '))
        rehab_budget = int(input('How much was your rehab budget? '))
        misc = int(input('Do you have any other miscellanous costs? '))
        
        self.investment = sum([down_payment, closing_costs, rehab_budget, misc])
        
        print(f' Total investment is {self.investment}')
        
        
        
    def calculate_roi(self):
        '''take annual cash flow and divide by total investment then multiply 
        by 100 to get ROI percentage'''
        
        self.cash_roi = self.annual_flow / self.investment * 100
        
        print(f'your total cash on cash return on investment is %{self.cash_roi:.2f}')
        
        
        
        
        
                
def calculate_roi():
    
    rental = ROIcalc()
    rental.monthly_income()
    rental.monthly_expense()
    rental.annual_cf()
    rental.total_investment()
    rental.calculate_roi()
        
        
        

        
    
```
